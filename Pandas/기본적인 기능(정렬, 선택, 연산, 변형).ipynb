## 변형

- df.groupby

```python
df = pd.read_csv("./python_data_basic-main/3장/CARD_SUBWAY_MONTH_202102.csv", index_col=False)
df.groupby('노선명').mean().sort_values(by='하차총승객수').head(5)
```

- pd.merge

```python
df1 = pd.DataFrame({'이름':['A','B','D'], '성별':['남','여','남']})
df2 = pd.DataFrame({'이름':['A','B','C'], '키':[180,160,170]})
df1.head()
```

```python
df2.head()
```

```python
pd.merge(left=df1, right=df2, how='left', on='이름') #left 데이터는 모두 다 포함 right, inner, outer 이용하면 됨
```

```python
pd.merge(left=df1, right=df2, how='left', left_on='이름', right_on='이름') #만약 두 컬럼 명이 다르다면 이런식으로
```

- pd.concat

```python
pd.concat([df1,df2], axis=0) #무지성 이어붙이기
```
