### Query문 활용

- 가독성과 편의성이 최대 장점

```python
import pandas as pd
df = pd.read_csv("./python_data_basic-main/3장/CARD_SUBWAY_MONTH_202102.csv") 
df.head()
```

- **논리연산자를 활용한 방법**
    - 이런식으로  ==, >, >=, <, <=, != 와 같은 논리연산자 모두 사용 가능하다. 
    - 마찬가지로 and(&)와 or(|) 연산 모두 사용 가능하다.

```python
df.query("사용일자 == '중앙선'").head()
```

- **in, not in을 활용한 방법**

```python
df.query("노선명 not in ['용문', '지평']").head()
```

- **index를 이용한 검색**

```python
df.query("index == 20210201")
```

```python
pd.Series(["나얼", "한혜진"]).str.
```

- **문자열 부분 검색**
    - df.query("컬럼명.str.contains("문자")")
    - 컬럼명.str.startswith("문자")
    - 컬럼명.str.endswith("문자")
    - 참고) Series 뒤에 str 붙여서 사용 가능 (Series.str.startswith() endswith()  etc)

```python
df.query("사용일자.str.contains('1')").head()
```

```python
df.query("사용일자.str.startswith('공항')").head()
```

- 컬럼별 데이터 비교

```python
df.query('승차총승객수 < 하차총승객수') #이런식으로 하차총승객수가 더 큰 데이터들만 나타난다
```

- **외부 변수(또는 함수) 참조 연산**
    - @를 이용해서 참조해줌

```python
find_list = ['용문','지평','아신']

df.query("노선명 in @find_list")
```
